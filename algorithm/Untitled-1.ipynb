```python
# Fixed dependencies - do not remove or change.
import pytest
import pandas as pd
import numpy as np
import datetime

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.impute import SimpleImputer

import matplotlib.pyplot as plt
import tensorflow as tf


print("loading data")
raw_data = pd.read_excel("hr.xlsx")
print("data loaded")

print("loading data")
raw_predict_data = pd.read_excel("hr_predict.xlsx")
print("data loaded")
```

```python

x = raw_data.iloc[:, :-1].values
y = raw_data.iloc[:, -1:].values
print(y[0])
for i in range(len(y)):
    if y[i] < 4:
        y[i] = 1.0
    else:
        y[i] = 0.0

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 3:])
x[:, 3:] = imputer.transform(x[:, 3:])




columnsToEncode = [0,1,2]
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), columnsToEncode)], remainder='passthrough')

x = ct.fit_transform(x).toarray()
#print(x[0])


#predictionLines = -14
#px = x[predictionLines:, :]
#py = y[predictionLines:]
#x = x[:predictionLines, :]
#y = y[:predictionLines]



x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
originalXTest = x_test


sc = StandardScaler()

x_train = sc.fit_transform(x_train)

x_test = sc.transform(x_test)



ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=18, activation="relu"))
ann.add(tf.keras.layers.Dense(units=18, activation="relu"))
ann.add(tf.keras.layers.Dense(units=18, activation="relu"))
ann.add(tf.keras.layers.Dense(units=18, activation="relu"))
ann.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
ann.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
ann.fit(x_train, y_train, batch_size = 32, epochs = 100)
```

```python
print(originalXTest[0])
```

```python

results = ann.predict(x_test)

resultsY = []

for result in results:
    if result > 0.9:
        answer = "TOP3"
    else:
        answer = "NOT3"
    resultsY.append(answer)

for i in range(0, len(resultsY)):
    print( int(originalXTest[i][849]), int(originalXTest[i][851]), y_test[i][0], resultsY[i], results[i][0])
```
